````python
import itertools as it

def permutations(string):
    perm_set = set(it.permutations(string))
    
    return list("".join(p) for p in perm_set)
````

````python
# Recursive solution
def permutations(string):
    if len(string) == 1:
        return set(string)
    
    first = string[0]
    rest = permutations(string[1:])
    result = set()
    
    for i in range(len(string)):
        for p in rest:
            result.add(p[0:i] + first + p[i:])
            
    return result
````
